```python
def compute_average_vector(words, model):
    vectors = []
    for word in words:
        if word in model.key_to_index:  # Check if the word is in the GloVe model
            vectors.append(model[word])
    if vectors:
        return np.mean(vectors, axis=0)  # Compute the average vector
    else:
        return np.zeros(model.vector_size)  # Return a zero vector if no words are found in the model
```

```python
def find_similar_to_average(average_vector, model, topn=100):
    # Find words similar to the average vector
    similar_words = model.similar_by_vector(average_vector, topn=topn)

    # Extract just the words from the returned tuples
    words = [word for word, similarity in similar_words]

    return words
```

### Once the curated and removed word lists have been developed, the following program will return the most semantically similar words to the ones on curated list and remove the words semantically similar to the removed words list. 

```python
def find_semantically_similar_words(curated_list, removed_list, topn=1000, similarity_threshold=0.5):
    # Compute average vectors for curated and removed lists
    average_vector_curated = compute_average_vector(curated_list, glove_model)
    average_vector_removed = compute_average_vector(removed_list, glove_model)
    
    # Find similar words to the curated list's average vector
    candidates = find_similar_to_average(average_vector_curated, glove_model, topn=topn)

    # Initialize a list to hold filtered candidates
    filtered_candidates = []

    for candidate in candidates:
        # Calculate similarity to the curated list's average
        sim_to_curated = np.dot(average_vector_curated, glove_model[candidate]) / (np.linalg.norm(average_vector_curated) * np.linalg.norm(glove_model[candidate]))
        
        # Calculate similarity to the removed list's average
        sim_to_removed = np.dot(average_vector_removed, glove_model[candidate]) / (np.linalg.norm(average_vector_removed) * np.linalg.norm(glove_model[candidate]))
        
        # Filter based on similarity thresholds
        if sim_to_curated > sim_to_removed and sim_to_removed < similarity_threshold:
            filtered_candidates.append(candidate)

    return filtered_candidates
```

### The below functions searches through the corpus of our candidates' tweets for any mention of the words on the curated list. It counts the tweets that have a hit and divides by the total sum tweets of the corpus. 
### It also records the index of the tweets for observation

```python
def analyze_word_presence(curated_list, tweets_mgp, tweets_deluzio, results_df):
    tokenized_curated_list = [preprocess_text(w)[0] for w in curated_list if preprocess_text(w)]

    relevant_tweets_mgp = []
    relevant_tweets_deluzio = []

    for idx, tweet in tweets_mgp.items():
        if any(w in tweet for w in tokenized_curated_list):
            relevant_tweets_mgp.append(idx)

    for idx, tweet in tweets_deluzio.items():
        if any(w in tweet for w in tokenized_curated_list):
            relevant_tweets_deluzio.append(idx)

    mgp_Proportion = len(relevant_tweets_mgp) / len(tweets_mgp)
    deluzio_Proportion = len(relevant_tweets_deluzio) / len(tweets_deluzio)

    # DataFrame for the current analysis
    result_df = pd.DataFrame({
        'Semantically_Similar_Words': [', '.join(curated_list)],
        'MGP_Proportion': [mgp_Proportion],
        'MGP_Relevant_Tweets': [relevant_tweets_mgp],
        'Deluzio_Proportion': [deluzio_Proportion],
        'Deluzio_Relevant_Tweets': [relevant_tweets_deluzio]
    })
    
    # Append the new results to the existing DataFrame and return
    updated_results_df = pd.concat([results_df, result_df], ignore_index=True)
    return updated_results_df
```

```python
def get_tweet_texts(tweet_indices, tweet_dataframe):
    # Retrieve 3 tweets from the index, if fewer than 3 exist include all
    sample_size = min(3, len(tweet_indices))
    np.random.seed(30)
    sampled_indices = np.random.choice(tweet_indices, size=sample_size, replace=False)
    tweets = tweet_dataframe.loc[sampled_indices, 'tweet_text_with_breaks'].tolist()
    # Concatenate tweet texts with a label before each tweet for clarity
    tweets_with_labels = [f"<strong>Ex. Tweet:</strong><br>{tweet}" for tweet in tweets]
    return "<br><br>".join(tweets_with_labels)

# Apply the function to each row and store the results in new columns
all_results_df['MGP_Tweet_Texts'] = all_results_df['MGP_Relevant_Tweets'].apply(lambda x: get_tweet_texts(x, mgp_tweets))
all_results_df['Deluzio_Tweet_Texts'] = all_results_df['Deluzio_Relevant_Tweets'].apply(lambda x: get_tweet_texts(x, deluzio_tweets))
```

```python
import pandas as pd

# Initialize lists to store data
topics = []
candidates = []
Proportions = []
semantically_similar_words = []

# Iterate over each row in the original DataFrame to populate the lists
for i, row in all_results_df.iterrows():
    first_word = row['Semantically_Similar_Words'].split(', ')[0]  # Extract the first word for simplicity
    # Add entry for MGP
    topics.append(first_word)
    candidates.append('MGP')
    Proportions.append(row['MGP_Proportion'] if 'MGP_Proportion' in all_results_df.columns else 0)  # Default to 0 if column doesn't exist
    semantically_similar_words.append(row['Semantically_Similar_Words'])

    # Add entry for Deluzio
    topics.append(first_word)
    candidates.append('Deluzio')
    Proportions.append(row['Deluzio_Proportion'] if 'Deluzio_Proportion' in all_results_df.columns else 0)  # Default to 0 if column doesn't exist
    semantically_similar_words.append(row['Semantically_Similar_Words'])

# Create the new DataFrame
simplified_df = pd.DataFrame({
    'Topic': topics,
    'Candidate': candidates,
    'Proportion': Proportions,
    'Semantically_Similar_Words': semantically_similar_words
})

# Display the new DataFrame
print(simplified_df.head())
```

```python
def insert_word_breaks(text, words_per_line=10):
    words = text.split(', ')
    # Insert a <br> tag after every 'words_per_line' words
    for i in range(words_per_line, len(words), words_per_line):
        words[i] = '<br>' + words[i]
    return ', '.join(words)

# Apply the function to the 'Semantically_Similar_Words' column
simplified_df['Semantically_Similar_Words'] = simplified_df['Semantically_Similar_Words'].apply(insert_word_breaks)
```

```python
# Function to map sample tweets based on the topic and candidate
def map_sample_tweets(row, all_results_df):
    # Find the first word of the topic as the key
    key = row['Topic'].split(',')[0]

    # Find the corresponding row in all_results_df
    corresponding_row = all_results_df[all_results_df['Semantically_Similar_Words'].str.contains(key, regex=False)]

    # Check if the current row in simplified_df is for MGP or Deluzio and assign the sample tweets accordingly
    if row['Candidate'] == 'MGP':
        return corresponding_row['MGP_Tweet_Texts'].iloc[0] if not corresponding_row.empty else "No tweets available"
    elif row['Candidate'] == 'Deluzio':
        return corresponding_row['Deluzio_Tweet_Texts'].iloc[0] if not corresponding_row.empty else "No tweets available"

# Apply the function to the simplified_df to create a new column for sample tweets
simplified_df['Sample_Tweets'] = simplified_df.apply(map_sample_tweets, axis=1, all_results_df=all_results_df)

# Display the updated DataFrame to confirm the sample tweets have been added
print(simplified_df.head())
```

```python
from bokeh.models import ColumnDataSource, FactorRange, CustomJSTickFormatter, HoverTool
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap

# Initialize the notebook output for Bokeh
output_notebook()

# Create a list of tuples for each combination of quoted topic and candidate
factors = [(quoted_topic, candidate) for quoted_topic in simplified_df['Quoted_Topic'].unique() for candidate in simplified_df['Candidate'].unique()]

# Update the ColumnDataSource to include a 'factors' column that combines 'Quoted_Topic' and 'Candidate'
simplified_df['factors'] = [(quoted_topic, candidate) for quoted_topic, candidate in zip(simplified_df['Quoted_Topic'], simplified_df['Candidate'])]
source = ColumnDataSource(simplified_df)

# Define colors for each candidate
candidate_colors = {'MGP': '#008080', 'Deluzio': '#7BAFD4'}

# Create the figure with the updated x_range using the new factors
p = figure(x_range=FactorRange(*factors), title="Candidate Tweet Proportions by Primary Topic",
           sizing_mode="inherit", tools="", toolbar_location=None)


# Draw the bars with proper color mapping and legend
p.vbar(x='factors', top='Proportion', width=0.9, source=source, legend_field='Candidate',
       fill_color=factor_cmap('Candidate', palette=list(candidate_colors.values()), factors=list(candidate_colors.keys())))


# Adjust title and axis labels
p.title.text_font_size = '16pt'
p.xaxis.major_label_orientation = 1
p.xaxis.major_label_text_font_size = "12pt"

hover = HoverTool(
    tooltips=[
        ("Candidate", "@Candidate"),
        ("Topic", "@Quoted_Topic"),
        ("Proportion", "@Proportion{0.0%}"),
        ("Words Used in Search", "@Semantically_Similar_Words{safe}"),
        ("Sample Tweets", "@Sample_Tweets{safe}")
    ],
    attachment="vertical"  # try "vertical", "above", or "below"
)
p.add_tools(hover)


# Aesthetic adjustments
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.formatter = CustomJSTickFormatter(code="return (tick*100).toFixed(0) + '%'")

# Display the plot
show(p)
```
